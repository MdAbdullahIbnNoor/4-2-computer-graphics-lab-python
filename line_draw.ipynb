```python
import matplotlib.pyplot as plt

def draw_line_DDA(x1, y1, x2, y2):
    # Calculate differences
    dx = x2 - x1
    dy = y2 - y1

    # Find the number of steps
    steps = max(abs(dx), abs(dy))

    # Calculate increments
    x_inc = dx / steps
    y_inc = dy / steps

    # Initialize starting point
    x, y = x1, y1

    # Plot the starting point
    plt.plot(round(x), round(y), marker='o', color='black')

    # Draw the line
    for _ in range(steps):
        x += x_inc
        y += y_inc
        plt.plot(round(x), round(y), marker='o', color='black')

    # Show the plot
    plt.title('DDA Line Drawing Algorithm')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()

# Example usage
draw_line_DDA(1, 2, 8, 10)
```

```python
import matplotlib.pyplot as plt

def draw_circle_midpoint(x_center, y_center, radius):
    x = radius
    y = 0
    p = 1 - radius  # Initial decision parameter for the midpoint

    # Collect coordinates in a list
    coordinates = [(x_center + x, y_center - y)]

    # Plot the initial points based on circle symmetry
    plot_symmetric_points(x_center, y_center, x, y, coordinates)

    while x > y:
        y += 1

        # Mid-point is inside or on the perimeter of the circle
        if p <= 0:
            p = p + 2*y + 1

        # Mid-point is outside the perimeter of the circle
        else:
            x -= 1
            p = p + 2*y - 2*x + 1

        # All the perimeter points have already been printed
        if x < y:
            break

        # Collect and plot the generated point in the other octants
        plot_symmetric_points(x_center, y_center, x, y, coordinates)

    # Plot the points
    plt.plot([point[0] for point in coordinates], [point[1] for point in coordinates], marker='o', linestyle='-', color='black')

    # Show the plot
    plt.title('Midpoint Circle Drawing Algorithm')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

def plot_symmetric_points(x_center, y_center, x, y, coordinates):
    # Plot the generated point and its reflection in the other octants
    coordinates.extend([
        (x_center + x, y_center - y),
        (x_center - x, y_center - y),
        (x_center + x, y_center + y),
        (x_center - x, y_center + y),
        (x_center + y, y_center - x),
        (x_center - y, y_center - x),
        (x_center + y, y_center + x),
        (x_center - y, y_center + x)
    ])

# Example usage
draw_circle_midpoint(0, 0, 5)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def boundary_fill(image, seed, fill_color, boundary_color):
    stack = [seed]

    while stack:
        x, y = stack.pop()

        if image[y, x] != fill_color and image[y, x] != boundary_color:
            image[y, x] = fill_color

            # Check and add neighboring pixels to the stack
            if x + 1 < image.shape[1]:
                stack.append((x + 1, y))  # right
            if x - 1 >= 0:
                stack.append((x - 1, y))  # left
            if y + 1 < image.shape[0]:
                stack.append((x, y + 1))  # up
            if y - 1 >= 0:
                stack.append((x, y - 1))  # down

# Example usage:
# Create a blank image with boundary and seed point
width, height = 10, 10
image = np.zeros((height, width), dtype=int)
image[4, 4] = 1  # Seed point
image[[0, -1], :] = 1  # Top and bottom boundary
image[:, [0, -1]] = 1  # Left and right boundary

# Display the initial image
plt.imshow(image, cmap='viridis', origin='upper', extent=[0, width, height, 0])
plt.title('Initial Image')
plt.show()

# Apply boundary fill algorithm
boundary_fill(image, (4, 4), 2, 1)

# Display the filled image
plt.imshow(image, cmap='viridis', origin='upper', extent=[0, width, height, 0])
plt.title('After Boundary Fill')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def add_edge(et, x1, y1, x2, y2):
    # Add edge to Edge Table (ET)
    if y1 > y2:
        x1, y1, x2, y2 = x2, y2, x1, y1
    et.append([y1, y2, x1, 1 / (x2 - x1), x1])

def update_aet(aet, y):
    # Update Active Edge Table (AET) at the current scanline
    for edge in aet:
        edge[4] += edge[3]
    aet = [edge for edge in aet if y < edge[1]]
    aet.sort(key=lambda x: x[4])
    return aet

def fill_scanline(aet, image, y):
    # Fill pixels between pairs of intersecting edges in AET
    for i in range(0, len(aet), 2):
        x1 = int(aet[i][4])
        x2 = int(aet[i + 1][4])
        for x in range(x1, x2 + 1):
            image[y, x] = 1  # Set pixel color (e.g., white)

def scanline_fill(poly_edges, image_height):
    et = []  # Edge Table
    aet = []  # Active Edge Table

    for edge in poly_edges:
        add_edge(et, *edge)

    for y in range(image_height):
        aet = update_aet(aet, y)
        fill_scanline(aet, image, y)

    plt.imshow(image, cmap='gray', origin='upper')
    plt.title('Scan-Line Fill Algorithm')
    plt.show()

# Example Usage:
image_width = 100
image_height = 100
image = np.zeros((image_height, image_width), dtype=int)

# Example polygon edges: (x1, y1, x2, y2)
polygon_edges = [(20, 30, 80, 90), (40, 10, 60, 70)]
scanline_fill(polygon_edges, image_height)
```

```python
import matplotlib.pyplot as plt

def cohen_sutherland(x1, y1, x2, y2, xmin, ymin, xmax, ymax):
    def compute_region_code(x, y):
        code = 0b0000
        if x < xmin:
            code |= 0b0001  # to the left of window
        elif x > xmax:
            code |= 0b0010  # to the right of window
        if y < ymin:
            code |= 0b0100  # below the window
        elif y > ymax:
            code |= 0b1000  # above the window
        return code

    code1 = compute_region_code(x1, y1)
    code2 = compute_region_code(x2, y2)

    while (code1 | code2) != 0:
        if (code1 & code2) != 0:
            # Line is completely outside the window
            return None, None, None, None
        else:
            # Clip the line
            if code1 != 0:
                if code1 & 0b0001:
                    x1 = xmin
                    y1 = y1 + (y2 - y1) * (xmin - x1) / (x2 - x1)
                elif code1 & 0b0010:
                    x1 = xmax
                    y1 = y1 + (y2 - y1) * (xmax - x1) / (x2 - x1)
                elif code1 & 0b0100:
                    y1 = ymin
                    x1 = x1 + (x2 - x1) * (ymin - y1) / (y2 - y1)
                elif code1 & 0b1000:
                    y1 = ymax
                    x1 = x1 + (x2 - x1) * (ymax - y1) / (y2 - y1)
                code1 = compute_region_code(x1, y1)
            else:
                if code2 & 0b0001:
                    x2 = xmin
                    y2 = y1 + (y2 - y1) * (xmin - x1) / (x2 - x1)
                elif code2 & 0b0010:
                    x2 = xmax
                    y2 = y1 + (y2 - y1) * (xmax - x1) / (x2 - x1)
                elif code2 & 0b0100:
                    y2 = ymin
                    x2 = x1 + (x2 - x1) * (ymin - y1) / (y2 - y1)
                elif code2 & 0b1000:
                    y2 = ymax
                    x2 = x1 + (x2 - x1) * (ymax - y1) / (y2 - y1)
                code2 = compute_region_code(x2, y2)

    return int(x1), int(y1), int(x2), int(y2)

# Example Usage:
xmin, ymin, xmax, ymax = 20, 20, 80, 80
x1, y1, x2, y2 = 10, 10, 90, 90

clipped_x1, clipped_y1, clipped_x2, clipped_y2 = cohen_sutherland(x1, y1, x2, y2, xmin, ymin, xmax, ymax)

if clipped_x1 is not None:
    plt.plot([x1, x2], [y1, y2], label='Original Line', color='blue')
    plt.plot([clipped_x1, clipped_x2], [clipped_y1, clipped_y2], label='Clipped Line', color='red')
    plt.xlim(0, 100)
    plt.ylim(0, 100)
    plt.axvline(xmin, color='gray', linestyle='--', label='Clipping Window')
    plt.axvline(xmax, color='gray', linestyle='--')
    plt.axhline(ymin, color='gray', linestyle='--')
    plt.axhline(ymax, color='gray', linestyle='--')
    plt.legend()
    plt.title('Cohen-Sutherland Line Clipping Algorithm')
    plt.show()
else:
    print('Line is completely outside the window.')
```

```python
import numpy as np
import matplotlib.pyplot as plt

def translation_translate(polygon, translate_x, translate_y):
	homogeneous_polygon = np.column_stack((polygon, np.ones(len(polygon))))
	translate_matrix = np.array([
			[1, 0, translate_x],
			[0, 1, translate_y],
			[0, 0, 1]
		])

	translated_polygon_homogeneous = np.dot(translate_matrix, homogeneous_polygon.T).T
	return translated_polygon_homogeneous[:, :2]

def translation_scaling(polygon, scale_x, scale_y):
	# Scaling matrix
	scaling_matrix = np.array([
	    [scale_x, 0],
	    [0, scale_y]
	])
	return np.dot(polygon, scaling_matrix.T)

def translation_reflection(polygon, x_axis=True, y_axis=True):
	x_axis = -1 if x_axis else 1
	y_axis = -1 if y_axis else 1
	# Scaling matrix
	scaling_matrix = np.array([
	    [x_axis, 0],
	    [0, y_axis]
	])
	return np.dot(polygon, scaling_matrix.T)

def translation_shearing(polygon, shear_x, shear_y):
	shear_matrix = np.array([
	    [1, shear_x],
	    [shear_y, 1]
	])

	return np.dot(polygon, shear_matrix.T) 

def translation_roation(polygon, angle_degrees):
	# Convert angle to radians
	angle_radians = np.radians(angle_degrees)

	# Rotation matrix
	rotation_matrix = np.array([
	    [np.cos(angle_radians), -np.sin(angle_radians)],
	    [np.sin(angle_radians), np.cos(angle_radians)]
	])
	return np.dot(polygon, rotation_matrix.T)




plt.figure(figsize=(9,6))

# Define the polygon (example: square)
polygon = np.array([(0, 0), (1, 0), (1, 1), (0, 1)])

plt.subplot(2,3,1)
plt.plot(polygon[:,0], polygon[:,1], c='black')
plt.plot(polygon[[0,-1]][0], polygon[[0,-1]][1], c='black')
plt.title('Original')


"""_______________________Translate_____________________________"""
translate_x = 3
translate_y = 3
translate_polygon = translation_translate(polygon, translate_x, translate_y)

plt.subplot(2,3,2)
plt.plot(translate_polygon[:,0], translate_polygon[:,1], c='red')
plt.plot(translate_polygon[[0,-1],0], translate_polygon[[0,-1],1], c='red')
plt.title('translate_polygon')
plt.xlim(min(np.min(polygon[:, 0]), np.min(translate_polygon[:, 0])) - 1,
         max(np.max(polygon[:, 0]), np.max(translate_polygon[:, 0])) + 1)
plt.ylim(min(np.min(polygon[:, 1]), np.min(translate_polygon[:, 1])) - 1,
         max(np.max(polygon[:, 1]), np.max(translate_polygon[:, 1])) + 1)

# Add labels and legend
plt.legend()



"""_______________________Scaling_____________________________"""
scale_x = 2
scale_y = 3
scaled_polygon = translation_scaling(polygon, scale_x, scale_y)
# Plot the original and scaled polygons
plt.subplot(2,3,3)
plt.plot(scaled_polygon[:,0], scaled_polygon[:,1], c='red')
plt.plot(scaled_polygon[[0,-1],0], scaled_polygon[[0,-1],1], c='red')

# Set plot limits
plt.xlim(min(np.min(polygon[:, 0]), np.min(scaled_polygon[:, 0])) - 1,
         max(np.max(polygon[:, 0]), np.max(scaled_polygon[:, 0])) + 1)
plt.ylim(min(np.min(polygon[:, 1]), np.min(scaled_polygon[:, 1])) - 1,
         max(np.max(polygon[:, 1]), np.max(scaled_polygon[:, 1])) + 1)

# Add labels and legend
plt.title('Scaled_polygon')
plt.legend()


"""_______________________reflect_____________________________"""
x_axis = True
y_axis = True
reflection_polygon = translation_reflection(polygon, x_axis=True, y_axis=True)
# Plot the original and scaled polygons
plt.subplot(2,3,4)
plt.plot(reflection_polygon[:,0], reflection_polygon[:,1], c='red')
plt.plot(reflection_polygon[[0,-1],0], reflection_polygon[[0,-1],1], c='red')

# Set plot limits
plt.xlim(min(np.min(polygon[:, 0]), np.min(reflection_polygon[:, 0])) - 1,
         max(np.max(polygon[:, 0]), np.max(reflection_polygon[:, 0])) + 1)
plt.ylim(min(np.min(polygon[:, 1]), np.min(reflection_polygon[:, 1])) - 1,
         max(np.max(polygon[:, 1]), np.max(reflection_polygon[:, 1])) + 1)

# Add labels and legend
plt.title('reflection_polygon_polygon')
plt.legend()



"""_______________________Shearing_____________________________"""
shear_x = 0.5  # Horizontal shear
shear_y = 0.2  # Vertical shear
shearing_polygon = translation_shearing(polygon, shear_x, shear_y)
# Plot the original and scaled polygons
plt.subplot(2,3,5)
plt.plot(shearing_polygon[:,0], shearing_polygon[:,1], c='red')
plt.plot(shearing_polygon[[0,-1],0], shearing_polygon[[0,-1],1], c='red')

# Set plot limits
plt.xlim(min(np.min(polygon[:, 0]), np.min(shearing_polygon[:, 0])) - 1,
         max(np.max(polygon[:, 0]), np.max(shearing_polygon[:, 0])) + 1)
plt.ylim(min(np.min(polygon[:, 1]), np.min(shearing_polygon[:, 1])) - 1,
         max(np.max(polygon[:, 1]), np.max(shearing_polygon[:, 1])) + 1)

# Add labels and legend
plt.title('shearing_polygon_polygon')
plt.legend()


"""_______________________rotation_____________________________"""
angle_degrees = 45
rotation_polygon = translation_roation(polygon, angle_degrees)
# Plot the original and scaled polygons
plt.subplot(2,3,6)
plt.plot(rotation_polygon[:,0], rotation_polygon[:,1], c='red')
plt.plot(rotation_polygon[[0,-1],0], rotation_polygon[[0,-1],1], c='red')

# Set plot limits
plt.xlim(min(np.min(polygon[:, 0]), np.min(rotation_polygon[:, 0])) - 1,
         max(np.max(polygon[:, 0]), np.max(rotation_polygon[:, 0])) + 1)
plt.ylim(min(np.min(polygon[:, 1]), np.min(rotation_polygon[:, 1])) - 1,
         max(np.max(polygon[:, 1]), np.max(rotation_polygon[:, 1])) + 1)

# Add labels and legend
plt.title('rotation_polygon_polygon')
plt.legend()





plt.show()














```

```python
import matplotlib.pyplot as plt
import numpy as np

def translate_point(point, dx, dy):
    x, y = point
    translated_point = (x + dx, y + dy)
    return translated_point

def translate_line(line, dx, dy):
    translated_line = [(x + dx, y + dy) for x, y in line]
    return translated_line

def translate_polygon(polygon, dx, dy):
    translated_polygon = [(x + dx, y + dy) for x, y in polygon]
    return translated_polygon

# Example Usage:
# Define points, lines, and polygons
point = (2, 3)
line = [(1, 1), (4, 4)]
polygon = [(3, 1), (5, 1), (5, 3), (3, 3)]

# Translation amounts
dx, dy = 2, 3

# Apply translation
translated_point = translate_point(point, dx, dy)
translated_line = translate_line(line, dx, dy)
translated_polygon = translate_polygon(polygon, dx, dy)

# Plot original and translated objects
plt.figure(figsize=(8, 8))

# Plot point
plt.subplot(2, 2, 1)
plt.scatter(*zip(*[point]), label='Original Point', color='blue')
plt.scatter(*zip(*[translated_point]), label='Translated Point', color='red')
plt.legend()
plt.title('Point Translation')

# Plot line
plt.subplot(2, 2, 2)
plt.plot(*zip(*line), label='Original Line', color='blue')
plt.plot(*zip(*translated_line), label='Translated Line', color='red')
plt.legend()
plt.title('Line Translation')

# Plot polygon
plt.subplot(2, 2, 3)
polygon.append(polygon[0])  # Close the polygon
plt.plot(*zip(*polygon), label='Original Polygon', color='blue')
plt.plot(*zip(*translated_polygon), label='Translated Polygon', color='red')
plt.legend()
plt.title('Polygon Translation')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def translate(polygon, dx, dy):
    return [(x + dx, y + dy) for x, y in polygon]

def scale(polygon, sx, sy):
    return [(x * sx, y * sy) for x, y in polygon]

def reflect(polygon, axis):
    if axis == 'x':
        return [(x, -y) for x, y in polygon]
    elif axis == 'y':
        return [(-x, y) for x, y in polygon]
    else:
        print("Invalid axis for reflection. Use 'x' or 'y'.")
        return polygon

def shear(polygon, shx, shy):
    return [(x + shx * y, y + shy * x) for x, y in polygon]

def rotate(polygon, angle_degrees):
    angle_radians = np.radians(angle_degrees)
    cos_theta = np.cos(angle_radians)
    sin_theta = np.sin(angle_radians)
    return [(x * cos_theta - y * sin_theta, x * sin_theta + y * cos_theta) for x, y in polygon]

# Example Usage:
original_polygon = [(1, 1), (2, 1), (2, 2), (1, 2)]

# Translation
translated_polygon = translate(original_polygon, 2, 3)

# Scaling
scaled_polygon = scale(original_polygon, 2, 0.5)

# Reflection
reflected_polygon_x = reflect(original_polygon, 'x')
reflected_polygon_y = reflect(original_polygon, 'y')

# Shearing
sheared_polygon = shear(original_polygon, 0.5, 0)

# Rotation
rotated_polygon = rotate(original_polygon, 45)

# Plotting
plt.figure(figsize=(10, 8))

# Original Polygon
plt.subplot(2, 3, 1)
polygon = original_polygon + [original_polygon[0]]  # Closing the polygon
plt.plot(*zip(*polygon), marker='o', label='Original Polygon')
plt.title('Original Polygon')

# Translated Polygon
plt.subplot(2, 3, 2)
polygon = translated_polygon + [translated_polygon[0]]
plt.plot(*zip(*polygon), marker='o', label='Translated Polygon')
plt.title('Translated Polygon')

# Scaled Polygon
plt.subplot(2, 3, 3)
polygon = scaled_polygon + [scaled_polygon[0]]
plt.plot(*zip(*polygon), marker='o', label='Scaled Polygon')
plt.title('Scaled Polygon')

# Reflected Polygon (X-axis)
plt.subplot(2, 3, 4)
polygon = reflected_polygon_x + [reflected_polygon_x[0]]
plt.plot(*zip(*polygon), marker='o', label='Reflected Polygon (X-axis)')
plt.title('Reflected Polygon (X-axis)')

# Reflected Polygon (Y-axis)
plt.subplot(2, 3, 5)
polygon = reflected_polygon_y + [reflected_polygon_y[0]]
plt.plot(*zip(*polygon), marker='o', label='Reflected Polygon (Y-axis)')
plt.title('Reflected Polygon (Y-axis)')

# Sheared Polygon
plt.subplot(2, 3, 6)
polygon = sheared_polygon + [sheared_polygon[0]]
plt.plot(*zip(*polygon), marker='o', label='Sheared Polygon')
plt.title('Sheared Polygon')

plt.tight_layout()
plt.show()
```
