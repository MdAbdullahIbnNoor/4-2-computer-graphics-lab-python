```python
import matplotlib.pyplot as plt

def draw_line_Bresenham(x1, y1, x2, y2):
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    
    # Decision parameter
    p = 2 * dy - dx
    
    # Determine the direction of the step
    if x1 < x2:
        x_step = 1
    else:
        x_step = -1

    if y1 < y2:
        y_step = 1
    else:
        y_step = -1

    # Initial point
    x, y = x1, y1

    # Plot the initial point
    plt.plot(x, y, marker='o', color='black')

    # Draw the line
    while x != x2:
        x += x_step
        if p < 0:
            p += 2 * dy
        else:
            y += y_step
            p += 2 * (dy - dx)
        plt.plot(x, y, marker='o', color='black')

    # Show the plot
    plt.title('Bresenham Line Drawing Algorithm')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()

# Example usage
draw_line_Bresenham(1, 2, 8, 10)
```

```python
import matplotlib.pyplot as plt

def draw_line_DDA(x1, y1, x2, y2):
    # Calculate differences
    dx = x2 - x1
    dy = y2 - y1

    # Find the number of steps
    steps = max(abs(dx), abs(dy))

    # Calculate increments
    x_inc = dx / steps
    y_inc = dy / steps

    # Initialize starting point
    x, y = x1, y1

    # Plot the starting point
    plt.plot(round(x), round(y), marker='o', color='black')

    # Draw the line
    for _ in range(steps):
        x += x_inc
        y += y_inc
        plt.plot(round(x), round(y), marker='o', color='black')

    # Show the plot
    plt.title('DDA Line Drawing Algorithm')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()

# Example usage
draw_line_DDA(1, 2, 8, 10)
```

```python
import matplotlib.pyplot as plt

def draw_line_Bresenham(x1, y1, x2, y2):
    dx = abs(x2 - x1)
    dy = abs(y2 - y1)
    
    # Decision parameter
    p = 2 * dy - dx
    
    # Determine the direction of the step
    if x1 < x2:
        x_step = 1
    else:
        x_step = -1

    if y1 < y2:
        y_step = 1
    else:
        y_step = -1

    # Initial point
    x, y = x1, y1

    # Collect coordinates in a list
    coordinates = [(x, y)]

    # Draw the line
    while x != x2:
        x += x_step
        if p < 0:
            p += 2 * dy
        else:
            y += y_step
            p += 2 * (dy - dx)
        coordinates.append((x, y))

    # Plot the points
    plt.plot([point[0] for point in coordinates], [point[1] for point in coordinates], marker='o', color='black')

    # Connect the points with a line
    plt.plot([point[0] for point in coordinates], [point[1] for point in coordinates], linestyle='-', color='black')

    # Show the plot
    plt.title('Bresenham Line Drawing Algorithm')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()

# Example usage
draw_line_Bresenham(1, 2, 8, 10)
```

```python
### DDF Algorithm
import matplotlib.pyplot as plt

def ddf_algo(x1, y1, x2, y2):
    
    dx = abs(x2-x1)
    dy = abs(y2-y1)
    
    step = max(dx, dy)
    
    x_inc = dx/step
    y_inc = dy/step
    
    x, y = x1, y1
    
    plt.plot(x, y, marker='o', color='black')
    
    for i in range(step):
        x += x_inc
        y += y_inc
        
        plt.plot(x, y, marker='o', color='black')
        # plt.plot(x, y, linestyle='-', color='black')
    
    plt.title('DDF Algorithm')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()
    
ddf_algo(1, 2, 8, 10)
```

```python
import matplotlib.pyplot as plt

def draw_circle_midpoint(x_center, y_center, radius):
    x = radius
    y = 0
    p = 1 - radius  # Initial decision parameter for the midpoint

    # Collect coordinates in a list
    coordinates = [(x_center + x, y_center - y)]

    # Plot the initial points based on circle symmetry
    plot_symmetric_points(x_center, y_center, x, y, coordinates)

    while x > y:
        y += 1

        # Mid-point is inside or on the perimeter of the circle
        if p <= 0:
            p = p + 2*y + 1

        # Mid-point is outside the perimeter of the circle
        else:
            x -= 1
            p = p + 2*y - 2*x + 1

        # All the perimeter points have already been printed
        if x < y:
            break

        # Collect and plot the generated point in the other octants
        plot_symmetric_points(x_center, y_center, x, y, coordinates)

    # Plot the points
    plt.plot([point[0] for point in coordinates], [point[1] for point in coordinates], marker='o', linestyle='-', color='black')

    # Show the plot
    plt.title('Midpoint Circle Drawing Algorithm')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

def plot_symmetric_points(x_center, y_center, x, y, coordinates):
    # Plot the generated point and its reflection in the other octants
    coordinates.extend([
        (x_center + x, y_center - y),
        (x_center - x, y_center - y),
        (x_center + x, y_center + y),
        (x_center - x, y_center + y),
        (x_center + y, y_center - x),
        (x_center - y, y_center - x),
        (x_center + y, y_center + x),
        (x_center - y, y_center + x)
    ])

# Example usage
draw_circle_midpoint(0, 0, 5)
```

```python
import matplotlib.pyplot as plt

def midPointCircleDraw(x_centre, y_centre, r):
    x = r
    y = 0
    
    # Collect coordinates in a list
    coordinates = [(x + x_centre, y + y_centre)]
    
    # Printing the initial point on the axes after translation
    print("(", x + x_centre, ", ", y + y_centre, ")", sep="", end="")
    
    # When radius is zero only a single point will be printed
    if r > 0:
        print("(", x + x_centre, ", ", -y + y_centre, ")", sep="", end="")
        print("(", y + x_centre, ", ", x + y_centre, ")", sep="", end="")
        print("(", -y + x_centre, ", ", x + y_centre, ")", sep="")
    
    # Initializing the value of P
    P = 1 - r
    
    while x > y:
        y += 1
        
        # Mid-point inside or on the perimeter
        if P <= 0:
            P = P + 2 * y + 1
        
        # Mid-point outside the perimeter
        else:
            x -= 1
            P = P + 2 * y - 2 * x + 1
        
        # All the perimeter points have already been printed
        if x < y:
            break
        
        # Collect and print the generated point and its reflection in the other octants after translation
        coordinates.extend([
            (x + x_centre, y + y_centre),
            (-x + x_centre, y + y_centre),
            (x + x_centre, -y + y_centre),
            (-x + x_centre, -y + y_centre),
        ])
        
        # If the generated point on the line x = y, then the perimeter points have already been printed
        if x != y:
            coordinates.extend([
                (y + x_centre, x + y_centre),
                (-y + x_centre, x + y_centre),
                (y + x_centre, -x + y_centre),
                (-y + x_centre, -x + y_centre),
            ])
    
    # Convert the coordinates to separate lists for x and y
    x_coords, y_coords = zip(*coordinates)
    
    # Plot the points
    plt.plot(x_coords, y_coords, marker='o', linestyle='', color='black')
    
    # Show the plot
    plt.title('Midpoint Circle Drawing Algorithm')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

# Example usage
if __name__ == '__main__':
    # To draw a circle of radius 3 centered at (0, 0)
    midPointCircleDraw(0, 0, 3)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def boundary_fill(image, seed, fill_color, boundary_color):
    stack = [seed]

    while stack:
        x, y = stack.pop()

        if image[y, x] != fill_color and image[y, x] != boundary_color:
            image[y, x] = fill_color

            # Check and add neighboring pixels to the stack
            if x + 1 < image.shape[1]:
                stack.append((x + 1, y))  # right
            if x - 1 >= 0:
                stack.append((x - 1, y))  # left
            if y + 1 < image.shape[0]:
                stack.append((x, y + 1))  # up
            if y - 1 >= 0:
                stack.append((x, y - 1))  # down

# Example usage:
# Create a blank image with boundary and seed point
width, height = 10, 10
image = np.zeros((height, width), dtype=int)
image[4, 4] = 1  # Seed point
image[[0, -1], :] = 1  # Top and bottom boundary
image[:, [0, -1]] = 1  # Left and right boundary

# Display the initial image
plt.imshow(image, cmap='viridis', origin='upper', extent=[0, width, height, 0])
plt.title('Initial Image')
plt.show()

# Apply boundary fill algorithm
boundary_fill(image, (4, 4), 2, 1)

# Display the filled image
plt.imshow(image, cmap='viridis', origin='upper', extent=[0, width, height, 0])
plt.title('After Boundary Fill')
plt.show()
```
